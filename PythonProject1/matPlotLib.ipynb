```python
import matplotlib.pyplot as plt
#绘制折线图
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

```python
import random
x = [i for i in range(10)]
y = [random.randint(0,10) for i in range(10)]
plt.plot(x,y)
plt.figure(figsize=(20,8),dpi=80)
plt.show()
```

```python
#绘制x,y轴刻度
x = [i for i in range(10)]
y = [random.randint(0,10) for i in range(10)]
plt.plot(x,y)
plt.xticks([i for i in range(0,10,2)])
plt.yticks([i for i in range(0,10,2)])
plt.figure(figsize=(20,8),dpi=80)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 核心：设置Matplotlib的字体为支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体：SimHei（黑体）
plt.rcParams['axes.unicode_minus'] = False    # 解决负号'-'显示为方块的问题

# 生成示例数据
x = np.arange(1, 7)
y = [12, 25, 18, 30, 22, 28]

# 绘图
plt.plot(x, y)
plt.xticks(ticks=x, labels=['一月', '二月', '三月', '四月', '五月', '六月'], rotation=45)
plt.title('月度销售额分析', fontsize=14)
plt.xlabel('月份', fontsize=12)
plt.ylabel('销售额（万元）', fontsize=12)

plt.tight_layout()
plt.show()

```
